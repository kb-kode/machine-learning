### Data Preparation

Our goal here is to break down each list of items in the products column into rows and count the number of products bought by a user

```python
# example 1: split product items
transactions['products'] = transactions['products'].apply(lambda x: [int(i) for i in x.split('|')])
transactions.head(2).set_index('customerId')['products'].apply(pd.Series).reset_index()
```

```python
# example 2: organize a given table into a dataframe with customerId, single productId, and purchase count
pd.melt(transactions.head(2).set_index('customerId')['products'].apply(pd.Series).reset_index(), 
             id_vars=['customerId'],
             value_name='products') \
    .dropna().drop(['variable'], axis=1) \
    .groupby(['customerId', 'products']) \
    .agg({'products': 'count'}) \
    .rename(columns={'products': 'purchase_count'}) \
    .reset_index() \
    .rename(columns={'products': 'productId'})
```

### Create data with user, item, and target field

This table will be an input for our modeling later
- In this case, our user is customerId, productId, and purchase_count

```python

s=time.time()

data = pd.melt(transactions.set_index('customerId')['products'].apply(pd.Series).reset_index(), 
             id_vars=['customerId'],
             value_name='products') \
    .dropna().drop(['variable'], axis=1) \
    .groupby(['customerId', 'products']) \
    .agg({'products': 'count'}) \
    .rename(columns={'products': 'purchase_count'}) \
    .reset_index() \
    .rename(columns={'products': 'productId'})
data['productId'] = data['productId'].astype(np.int64)

print("Execution time:", round((time.time()-s)/60,2), "minutes")
```

```python
product_list = list(set(transactions['products'].apply(lambda x:x.split('|')).sum()))
```

```python
product_list = sorted([int(x) for x in product_list])
```

```python
#normalize purchase frequency of each item across users by first creating a user-item matrix
user_item_count = {}

for c in customers:
    #Create empty list to fill for each user's pruchase history
    id_list = [0 for i in range(0, max(product_list)+1)]
    
    customer_purchases = transactions[transactions['customerId'] == c]['products']
    
    for purchase in customer_purchases:
        for i in purchase.split('|'):
            id_list[int(i)] += 1
    user_item_count[c] = id_list
```

```python
user_item_df = pd.DataFrame(user_item_count).T
user_item_df.index.names = ['customerId']
user_item_df.replace(0, np.nan, inplace=True)
```

```python

user_item_df_norm = (user_item_df-user_item_df.min())/(user_item_df.max()-user_item_df.min())
```

```python
# create a table for input to the modeling
d = user_item_df_norm.reset_index() 
d.index.names = ['scaled_purchase_freq'] 
```

```python
data_norm = pd.melt(d, id_vars=['customerId'], value_name='scaled_purchase_freq')
```

```python
data_norm.columns = ['customerId','productId','scaled_purchase_freq']
```

```python
data_norm.dropna(inplace=True)
```
